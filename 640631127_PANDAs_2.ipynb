# Getting Data from other resources

We can get the data from any web services as well
See the example of getting the [Bank Of Thailand example](https://apiportal.bot.or.th/bot/public/)

```
import requests


url = "https://apigw1.bot.or.th/bot/public/DepositRate/v2/deposit_rate_/"
headers = {
    'x-ibm-client-id': "f065ef48-c8d3-4392-a79a-5387789a78da",
    'accept': "application/json"
    }

resp = requests.request("GET", url, headers=headers, params={"start_period":"2021-08-01", "end_period":"2021-08-31"})
```

```
res_j = pd.io.json.loads(resp.text)
print(type(res_j))
res_j
```
