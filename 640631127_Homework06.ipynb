# Import Salaries Dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("avocado.csv")
df.head(50)
```

# Using Python Code to find:
## 1) Which region sold the largest amount of avocado ?

```python
def biggest(region,column):
    df_group = df.groupby([region])[[column]].sum()
    df_sorted = df_group.sort_values( by = column, ascending = False)
    df_sorted[column] = df_sorted[column].astype('int64') 
    df_sorted.head()
    return df_sorted.head()
```

```python
biggest('region','Total Volume')
```

Answer TotalUS sold the largest amount of avocado.

Answer TotalUS has the biggest lot of sold avocado came from 4225 with 2015011554 units.

```python
biggest('region','4770')
```

# 2)Which region sold the smallest amount of avocado ?

```python
def smallest(region,column):
    df_group = df.groupby([region])[[column]].sum()
    df_sorted = df_group.sort_values( by = column, ascending = True)
    df_sorted[column] = df_sorted[column].astype('int64') 
    df_sorted.head()
    return df_sorted.head()
```

```python
smallest('region','Total Volume')
```

# 3) Which region sold the highest price of avocado in average ?

```python
df_group = df.groupby(['region'])[['AveragePrice']].mean()
df_sorted = df_group.sort_values( by = 'AveragePrice', ascending = False)
df_sorted.head()
```

Answer HartfordSpringfield sold the highest price of avocado in average with 1.818639 per unit

# 4) Find the total amount of income (Avg_Price*Total_Volume) of each region.

```python
df['Income'] = df['AveragePrice'] * df['Total Volume']
df_group = df.groupby(['region'])[['Income']].sum()
df_sorted = df_group.sort_values( by = 'Income', ascending = False)
df_sorted['Income'] = df_sorted['Income'].astype('int64') 
df_sorted
```

So, the total amount of income (Avg_Price * Total_Volume) of each region as above.

# 5) Let AVOCADO  Average Weight : <br>4046 => 4 ounces, <br>4225 => 9 ounces, <br>4770 => 12 ounces<br><br>5.1)Find the number of sold avocadoes by region ?

```python
df['AVOCADO Weight'] = df['4046'] * 4 + df['4225'] * 9 + df['4770'] * 12
df_group = df.groupby(['region'])[['AVOCADO Weight']].sum()
df_sorted = df_group.sort_values( by = 'AVOCADO Weight', ascending = False)
df_sorted['AVOCADO Weight'] = df_sorted['AVOCADO Weight'].astype('int64') 
df_sorted
```

So, the number of sold avocadoes by region as above.

# 6) Normally, the customers buy the avocados by unit or in a bags ?

```python
df['Total units'] = df['4046'] + df['4225'] + df['4770']
df = df.groupby(['region'])[['Total units','Total Bags']].sum()
df = df[['Total units','Total Bags']].astype('int64') 
df
```

```python
df['Normally buy'] = np.where(df['Total units'] > df['Total Bags'], 'by units', 'by bags')
df
```

```python
df[['Normally buy']].value_counts(normalize=True)
```

Answer Normally, the customers buy the avocados by unit by 98.1%
